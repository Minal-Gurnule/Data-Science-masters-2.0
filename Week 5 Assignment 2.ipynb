```python
''' Que 3 The arithmetic error occurs when an error is encountered during numeric calculations in Python. 
This includes 
Zero division Error 
Floating point error
Over Flow error
In addition, zero division error is raised when you divide a numeric value by zero. When we run this code, we will get a 
ZeroDivision error.'''
```

```python
# Zero division Error 

try:
    a = 10
    b = 0
    print(a/b)
    
except ZeroDivisionError as e:
    print("Error: Cannot divide by zero")
```

```python
# Over Flow error

j = 5.0

try:
    for i in range(1, 10):
        j = j**i
        print(j)
        
except OverflowError as e:
    print("Overflow error happened")
```

```python
''' Que 4 LookupError Exception is the Base class for errors raised when something can't be found. The base class for the 
exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError. An IndexError 
is raised when a sequence reference is out of range.'''
```

```python
# Key Error

person_info = {'name': 'Lenin Mishra','age': 28,'language': 'Python'}

person_input = input('What do you want to learn about Person==> ')

try:
    print(f'{person_input} is {person_info[person_input]}')
    
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

```python
# Index Error

x = [1, 2, 3, 4]
try:
    print(x[5])
    
except LookupError as e:
    print(f"{e}, {e.__class__}")
```
